```python
class Graph:
    def __init__(self):
        self.graph = {}

    def add_edge(self, u, v):
        if u not in self.graph:
            self.graph[u] = []
        self.graph[u].append(v)

    def print_graph(self):
        for vertex in self.graph:
            print(vertex, "->", " -> ".join(map(str, self.graph[vertex])))

g = Graph()
g.add_edge(1, 2)
g.add_edge(1, 3)
g.add_edge(1, 9)
g.add_edge(2, 1)
g.add_edge(2, 4)
g.add_edge(3, 1)
g.add_edge(3, 4)
g.add_edge(3, 5)
g.add_edge(3, 7)
g.add_edge(4, 2)
g.add_edge(4, 6)
g.add_edge(5, 3)
g.add_edge(5, 6)
g.add_edge(6, 4)
g.add_edge(6, 5)
g.add_edge(6, 7)
g.add_edge(8, 7)
g.add_edge(8, 9)
g.add_edge(9, 8)
g.add_edge(9, 1)

g.print_graph()
```

```python
class Graph:
    def __init__(self, num_vertices):
        self.num_vertices = num_vertices
        self.adj_matrix = [[0] * num_vertices for _ in range(num_vertices)]

    def add_edge(self, u, v):
        self.adj_matrix[u - 1][v - 1] = 1
        self.adj_matrix[v - 1][u - 1] = 1

    def print_graph(self):
        for row in self.adj_matrix:
            print(" ".join(map(str, row)))

    def print_adjacency_matrix(self):
        for row in self.adj_matrix:
            print(" ".join(map(str, row)))


g = Graph(9)
g.add_edge(1, 2)
g.add_edge(1, 3)
g.add_edge(1, 9)
g.add_edge(2, 1)
g.add_edge(2, 4)
g.add_edge(3, 1)
g.add_edge(3, 4)
g.add_edge(3, 5)
g.add_edge(3, 7)
g.add_edge(4, 2)
g.add_edge(4, 6)
g.add_edge(5, 3)
g.add_edge(5, 6)
g.add_edge(6, 4)
g.add_edge(6, 5)
g.add_edge(6, 7)
g.add_edge(8, 7)
g.add_edge(8, 9)
g.add_edge(9, 8)
g.add_edge(9, 1)

g.print_adjacency_matrix()
```
