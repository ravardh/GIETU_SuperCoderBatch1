```python
class Job:
    def __init__(self, profit, deadline):
        self.profit = profit
        self.deadline = deadline

# Define Jobs and n
Jobs = [Job(50, 2), Job(20, 1), Job(40, 1), Job(30, 1), Job(10, 3)]
n = len(Jobs)

# Define the job_scheduling function
def job_scheduling(Jobs, n):
    max_deadline = -1
    job_list = []

    for job in Jobs:
        max_deadline = max(max_deadline, job.deadline)
        job_list.append([job.profit, job.deadline])

    job_list.sort(reverse=True)

    job_slots = [-1] * (max_deadline + 1)

    for job in job_list:
        deadline = job[1]
        while deadline > 0 and job_slots[deadline] != -1:
            deadline -= 1
        if deadline >= 0:
            job_slots[deadline] = job[0]

    job_slots = job_slots[1:]

    unassigned_jobs = job_slots.count(-1)
    total_profit = sum([profit for profit in job_slots if profit != -1]) + unassigned_jobs

    return [max_deadline - unassigned_jobs, total_profit]

# Example usage:
result = job_scheduling(Jobs, n)
print(result)
```
