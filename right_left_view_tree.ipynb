```python
# right view
class Node:
    def __init__(self, item):
        self.data = item
        self.left = None
        self.right = None

def right_view_util(root, level, max_level_reached):
    if root is None:
        return

    if level > max_level_reached[0]:
        print(f"At level {level}: {root.data}")
        max_level_reached[0] = level

    right_view_util(root.right, level + 1, max_level_reached)
    right_view_util(root.left, level + 1, max_level_reached)

def right_view(root):
    max_level_reached = [0]
    print("Right View of the Binary Tree:")
    right_view_util(root, 1, max_level_reached)

# Create the binary tree
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)
root.right.left.right = Node(8)

# Display the right view of the binary tree
right_view(root)
```
