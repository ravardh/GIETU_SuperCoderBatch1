```python
class node:
    def __init__(self,data=None):
        self.data=data
        self.left=None
        self.right=None

def insert_into_bst(root,data):
    if root==None:
        return node(data)
  
    if data < root.data:
        root.left = insert_into_bst(root.left,data)
    else:
        root.right = insert_into_bst(root.right,data)
    return root    

def preorder(root):
    if root==None:
        return
    print(root.data, end=" ")
    preorder(root.left)
    preorder(root.right)
    
if __name__ == "__main__":
    arr = [10,14,17,9,5,15,12,7,8,2,3]
    root = None
    for x in range (0,len(arr)):
        root = insert_into_bst(root, arr[x]) 
print("\n\npreorder")
preorder(root)
```

```python
#another way(recursion)
def insertion(root,data):
    if data < root.data:
        if(root.left==None):
            root.left=node(data)
            return
        insertion(root.left,data)
    elif:
        if(root.right==None):
            root.right=node(data)
            return
        insertion(root.right,data)
```
