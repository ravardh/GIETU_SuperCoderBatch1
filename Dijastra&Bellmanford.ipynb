```python
class Graph:
    def __init__(self, vertices):
        self.graph = {}
        self.V = vertices

    def add_edge(self, u, v, w):
        if u not in self.graph:
            self.graph[u] = {}
        self.graph[u][v] = w

        if v not in self.graph:
            self.graph[v] = {}
        self.graph[v][u] = w

    def create_graph(self, num_vertices):
        for i in range(num_vertices):
            self.graph[i] = {}

    def print_graph(self):
        print("{")
        for i in self.graph:
            print(f"  {i}: {self.graph[i]}")
        print("}")

    def dijkstra(self, start):
        d = {}
        path = {}
        for i in self.graph:
            path[i] = float('inf')
        path[start] = 0
        k = start

        for i in self.graph:
            d[i] = float('inf')

        while len(d.keys()):
            v = path[k]
            for i in self.graph[k]:
                if i in d:
                    d[i] = min(d[i], v + self.graph[k][i])
                    path[i] = d[i]
            del d[k]
            if len(d):
                k = min(d, key=d.get)

        for i in self.graph:
            print(f"Minimum Distance of vertice {i} from {start} is {path[i]}")


g = Graph(9)
g.create_graph(10)
g.add_edge(1, 2, 4)
g.add_edge(1, 8, 8)
g.add_edge(2, 3, 8)
g.add_edge(2, 8, 11)
g.add_edge(3, 4, 7)
g.add_edge(3, 6, 4)
g.add_edge(3, 9, 2)
g.add_edge(4, 5, 9)
g.add_edge(4, 6, 14)
g.add_edge(5, 6, 10)
g.add_edge(6, 7, 2)
g.add_edge(7, 8, 1)
g.add_edge(7, 9, 6)
g.add_edge(8, 9, 7)

print(g.print_graph())

start = 1
g.dijkstra(start)
```

```python
# belllman for -ve weight values
class Graph:
    def __init__(self, vertices):
        self.graph = {}
        self.V = vertices

    def add_edge(self, u, v, w):
        if u not in self.graph:
            self.graph[u] = {}
        self.graph[u][v] = w

    def print_graph(self):
        print("{")
        for i in self.graph:
            print(f"  {i}: {self.graph[i]}")
        print("}")

    def bellman_ford(self, start):
        edge_list = [(i, j) for i in self.graph for j in self.graph[i]]

        path = {i: float('inf') for i in self.graph}
        path[start] = 0

        for _ in range(self.V - 1):
            for u, v in edge_list:
                path[v] = min(path[v], path[u] + self.graph[u][v])

        for i in self.graph:
            print(f"Minimum Distance of vertex {i} from {start} is {path[i]}")

g = Graph(7)
g.add_edge(1, 2, 6)
g.add_edge(1, 3, 5)
g.add_edge(1, 4, 5)
g.add_edge(2, 5, -1)
g.add_edge(3, 2, -2)
g.add_edge(3, 5, 1)
g.add_edge(4, 3, -2)
g.add_edge(4, 6, -1)
g.add_edge(5, 7, 3)
g.add_edge(6, 7, 3)
g.add_edge(7, 7, 0)

g.print_graph()
print()
g.bellman_ford(1)
```
