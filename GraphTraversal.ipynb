```python
# BFS traversal for graph having certain weight values attached to the graph
def create_graph(num_vertices):
    graph = {}
    for i in range(num_vertices):
        graph[i] = []
    return graph

def add_edge(graph, u, v, w):
    graph[u].append((v, w))

def bfs(graph, source):
    visited = [False for _ in range(len(graph))]
    queue = []
    visited[source] = True
    queue.append(source)

    while queue:
        current = queue.pop(0)
        print(current, end=" -> ")

        for neighbor, weight in graph[current]:
            if not visited[neighbor]:
                visited[neighbor] = True
                queue.append(neighbor)

# Example usage
graph = create_graph(5)
add_edge(graph, 0, 1, 4)
add_edge(graph, 0, 2, 8)
add_edge(graph, 1, 2, 3)
add_edge(graph, 1, 3, 6)
add_edge(graph, 2, 3, 7)
add_edge(graph, 2, 4, 10)
add_edge(graph, 3, 4, 9)
print_graph(graph)
print("BFS Traversal:")
bfs(graph, 0)
```

```python
# BFS traversal for UNWEIGHTED GRAPHS
def create_graph(num_vertices):
    graph = {}
    for i in range(num_vertices):
        graph[i] = []
    return graph

def add_edge(graph, u, v, weight=0):
    graph[u].append((v, weight))
    graph[v].append((u, weight))

def bfs(graph, source):
    visited = set()
    queue = [(source, 0)]
    while queue:
        node, distance = queue.pop(0)
        if node not in visited:
            visited.add(node)
            for neighbor, weight in graph[node]:
                if neighbor not in visited:
                    queue.append((neighbor, distance + weight))
    return visited

def print_graph(graph):
    for node in graph:
        print(node, graph[node])

# Example usage
num_vertices = 5
graph = create_graph(num_vertices)
add_edge(graph, 0, 1)
add_edge(graph, 0, 2)
add_edge(graph, 1, 3)
add_edge(graph, 2, 4)
print_graph(graph)
print("BFS Traversal:")
visited = bfs(graph, 0)
for node in visited:
    print(node, end=" ")
```
