```python
#insertion of elements in the bst
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

def insert(root, key):
    if root is None:
        return Node(key)
    else:
        if root.val < key:
            root.right = insert(root.right, key)
        else:
            root.left = insert(root.left, key)
    return root

def inorder_traversal(root):
    if root:
        inorder_traversal(root.left)
        print(root.val)
        inorder_traversal(root.right)

# Input array of elements
elements = [5, 3, 7, 2, 4, 6, 8]

# Construct the binary search tree
root = Node(elements[0])
for element in elements[1:]:       #insertion starts from the second element as 1st considered as root
    insert(root, element)

# Print the inorder traversal of the binary search tree
print("The BST formed is:")
inorder_traversal(root)
```

```python
# insertion of elements using tradotional approach
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

def insert(root, key):
    current = root
    while current:
        if key < current.val:
            if current.left is None:
                current.left = Node(key)
                break
            else:
                current = current.left
        else:
            if current.right is None:
                current.right = Node(key)
                break
            else:
                current = current.right

def construct_bst(arr):
    if not arr:
        return None

    root = Node(arr[0])

    for i in range(1, len(arr)):
        insert(root, arr[i])

    return root

def inorder_traversal(node):
    res = []
    stack = []
    current = node

    while True:
        if current is not None:
            stack.append(current)
            current = current.left
        elif stack:
            current = stack.pop()
            res.append(current.val)
            current = current.right
        else:
            break

    return res

# Input array
arr = [7, 4, 9, 1, 6, 8, 10]

# Construct the binary search tree
root_node = construct_bst(arr)

# Inorder traversal to print the elements in sorted order
result = inorder_traversal(root_node)
print(result)
```
