```python
# Python code for creating an adjacency list for an inputted graph

# Function to create an adjacency list for a graph
def create_adjacency_list(graph):
    adjacency_list = {}

    for edge in graph:
        if edge[0] in adjacency_list:
            adjacency_list[edge[0]].append(edge[1])
        else:
            adjacency_list[edge[0]] = [edge[1]]

        if edge[1] in adjacency_list:
            adjacency_list[edge[1]].append(edge[0])
        else:
            adjacency_list[edge[1]] = [edge[0]]

    return adjacency_list

# Input the graph as a list of edges
graph = [(0, 1), (0, 2), (1, 2), (2, 3), (3, 4)]

# Create the adjacency list for the inputted graph
adj_list = create_adjacency_list(graph)

# Print the adjacency list
print("The adjacency list is:")
print(adj_list)
```

```python
# Python code for creating an adjacency list from a user-inputted graph with a user-inputted number of nodes

def create_graph(graph, source, destination, num_nodes):
    if source >= num_nodes or destination >= num_nodes:
        return "Invalid source or destination node."

    graph[source].append(destination)
    graph[destination].append(source)

# Taking user input for the number of nodes in the graph
num_nodes = int(input("Enter the number of nodes in the graph: "))

# Initializing an empty adjacency list
graph = [[] for _ in range(num_nodes)]

# Taking user input for the edges of the graph
edges = int(input("Enter the number of edges: "))
for _ in range(edges):
    source = int(input("Enter the source node: "))
    destination = int(input("Enter the destination node: "))
    create_graph(graph, source, destination, num_nodes)

# Printing the adjacency list
for node in range(num_nodes):
    print(f"Node {node}:", end=" ")
    for neighbor in graph[node]:
        print(neighbor, end=" ")
    print()
    print(node, graph[node])
```
