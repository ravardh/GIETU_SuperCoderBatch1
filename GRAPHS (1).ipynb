```python
#creating adjacency list and matrix for storing the graph
def create_graph(num_vertices):
    graph = {}
    adjacency_matrix = [[0 for _ in range(num_vertices)] for _ in range(num_vertices)]
    for i in range(num_vertices):
        graph[i] = []
    return graph, adjacency_matrix

def add_edge(graph, adjacency_matrix, u, v):
    graph[u].append(v)
    graph[v].append(u)
    adjacency_matrix[u][v] = 1
    adjacency_matrix[v][u] = 1

def print_graph(graph, adjacency_matrix):
    print("Adjacency List:")
    for node in graph:
        print(node, graph[node])
    print("Adjacency Matrix:")
    for i in range(len(graph)):
        for j in range(len(graph)):
            print(adjacency_matrix[i][j], end=" ")
        print()

# Example usage
num_vertices = 5
graph, adjacency_matrix = create_graph(num_vertices)
add_edge(graph, adjacency_matrix, 0, 1)
add_edge(graph, adjacency_matrix, 0, 2)
add_edge(graph, adjacency_matrix, 1, 3)
add_edge(graph, adjacency_matrix, 2, 4)
print_graph(graph, adjacency_matrix)
```
