```python
class Linked_list:
    class Node:
        def __init__(self,data=None):
            self.data=data
            self.next=None
    def __init__(self):
        self.head=None
    def creation(self):
        print("1.Insert at beginning\n2.Insert at end")
        x=int(input())
        match(x):
            case 1:
                data=int(input("Enter data"))
                new_node=self.Node(data)
                if self.head is None:
                    self.head=new_node
                else:
                    new_node.next=self.head
                    self.head=new_node
            case 2:
                data=int(input("Enter data"))
                new_node=self.Node(data)
                if self.head is None:
                    self.head=new_node
                else:
                    last=self.head
                    while last.next is not None:
                        last=last.next
                    last.next=new_node
    def deletion(self):
        if self.head is None:
            print("List is empty")
        else:
            print("1.Delete at beginning\n2.Delete from end\n3.Delete at particular position")
            d=int(input())
            match(d):
                case 1:
                    temp=self.head
                    temp=temp.next
                    self.head=temp
                case 2:
                    temp=self.head
                    prev=temp.next
                    while(prev.next is not None):
                        temp=prev
                        prev=prev.next
                    temp.next=None
                case 3:
                    pos=int(input("Enter position"))
                    temp=self.head
                    prev=temp.next
                    for i in range(0,pos-2):
                        temp=prev
                        prev=prev.next
                    temp.next=prev.next
    def traversal(self):
        temp=self.head
        while temp is not None:
            print(temp.data,end=" ")
            temp=temp.next
l=Linked_list()
n=1
while(n==1):
    print("1.Insertion\n2.Deletion\n3.Print Linked List\n4.stop")
    x=int(input("Enter operation"))
    match(x):
        case 1:
            d=1
            while(d==1):
                l.creation()
                d=int(input("Enter more data:press 1"))
            n=int(input("Enter 1 to continue:"))
        case 2:
            d=1
            while(d==1):
                l.deletion()
                d=int(input("Want to delete more:press1"))
            n=int(input("Enter 1 to continue:"))
        case 3:
            print("Linked list is:")
            l.traversal()
            n=int(input("Enter 1 to continue:"))
        case 4:
            break
        case _:
            print("Enter valid operation:")
            n=int(input("Enter 1 to continue:"))
```
